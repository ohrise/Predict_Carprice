```python
import numpy as np

# 데이터 필터링
filtered_data = train[train['모델출시년도'] - train['생산년도'] > 2]
year_diff = filtered_data['모델출시년도'] - filtered_data['생산년도']

# IQR 계산
Q1 = np.percentile(year_diff, 25)
Q3 = np.percentile(year_diff, 75)
IQR = Q3 - Q1

# 이상치 조건 정의
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
outliers = year_diff[(year_diff < lower_bound) | (year_diff > upper_bound)]

print("이상치 개수:", len(outliers))
print("이상치 데이터:\n", outliers)
```

```python
import pandas as pd

# 배기량과 주행거리의 IQR을 구하기 위한 데이터 예시 (train DataFrame에 있다고 가정)
displacement = train['배기량']
mileage = train['주행거리']

# 배기량 이상치 찾기
Q1_displacement = displacement.quantile(0.25)
Q3_displacement = displacement.quantile(0.75)
IQR_displacement = Q3_displacement - Q1_displacement

lower_bound_displacement = Q1_displacement - 1.5 * IQR_displacement
upper_bound_displacement = Q3_displacement + 1.5 * IQR_displacement

# 주행거리 이상치 찾기
Q1_mileage = mileage.quantile(0.25)
Q3_mileage = mileage.quantile(0.75)
IQR_mileage = Q3_mileage - Q1_mileage

lower_bound_mileage = Q1_mileage - 1.5 * IQR_mileage
upper_bound_mileage = Q3_mileage + 1.5 * IQR_mileage

# 이상치인 데이터
displacement_outliers = train[(displacement < lower_bound_displacement) | (displacement > upper_bound_displacement)]
mileage_outliers = train[(mileage < lower_bound_mileage) | (mileage > upper_bound_mileage)]

print(f"배기량 이상치 개수: {len(displacement_outliers)}")
print(f"주행거리 이상치 개수: {len(mileage_outliers)}")
```
